# Total Parks in Pittsburgh Neighborhoods

Looking into the number of parks a neighborhood has can showcase the quality of life, environmental awareness, urban planning, and economic development.

## Why Total Parks?

Parks and green spaces are essential components of a healthy and vibrant community. They provide opportunities for recreation, relaxation, and social interaction and help enhance residents' overall quality of life. In Pittsburgh, a city known for its hilly terrain and abundance of trees, parks play a significant role in the community. By examining the number of parks in each Pittsburgh neighborhood, we can gain insights into the city's overall level of green space and assess how well each area is equipped to provide its residents with access to these essential resources.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Parks.csv")
data.sample(20)
```

----

## Mapping

```python
import matplotlib.pyplot as plt

plt.scatter(data['longitude'], data['latitude'])
plt.title('Parks in Pittsburgh')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
from PIL import Image

img = Image.open('Pitt.jpeg')

new_size = (600, 450)
img = img.resize(new_size)

img.show()
```

The scatterplot is a helpful tool to help visualize the distribution of parks in various neighborhoods in Pittsburgh. Plotting each park's longitude and latitude coordinates on a map allows us to see the concentration of parks in different areas of the city.

----

## Top Ten Neighborhoods

```python
category = data['neighborhood'].value_counts()

category.plot(kind='bar')

plt.title('Park Counts')
plt.xlabel('Neighborhoods')
plt.ylabel('Count')

plt.show()
```

```python
parks_neighborhood = data.groupby('neighborhood')['name'].count()
parks_neighborhood = parks_neighborhood.sort_values(ascending=False)

top_neighborhoods = parks_neighborhood.head(73)
counter = 0
prev_count = -1

for neighborhood, count in top_neighborhoods.items():
    if count != prev_count:
        counter += 1
    print(f'{counter}. {neighborhood}: {count} parks')
    prev_count = count
```
